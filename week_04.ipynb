### nltk
- Frequency Distribution: 단어가 몇 번 출현했는지 (빈도수)
- 텍스트의 성질 알기 위해
    - 단어 개수
    - 단어 빈도수
- FreqDist(text)

```python
import nltk
from nltk.book import *
nltk.download('book', quiet=True)
```

```python
fdist = FreqDist(text1) # cased
```

```python
fdist # 딕셔너리 형태 (,가 18713번 사용됨, ...)
```

```python
fdist.most_common(10) # 가장 빈번하게 쓰인 단어(부호포함)를 순서대로 출력
```

```python
fdist['The'] # 특정 단어의 출현 빈도수 출력
```

```python
fdist['that'], text1.count('that')
```

```python
fdist.freq('that') # 출현율 출력
```

```python
fdist.plot(50, cumulative = True) # 단어의 출현 빈도수를 그림으로 표현 (fdist.plot(단어 개수, cumulative = True))
```

```python
# 가장 빈번하지 않은(1번 출현), 희귀한 단어들
set(fdist.hapaxes()[:5])
```

### 리스트 <-> 스트링

    - default = ' '

```python
sent = 'I, have, a, car'
# ['I', 'have', 'a', 'car'] # tokens; sent를 리스트로
sent.split(',') # 쉼표로 구분
```

```python
# 공백 없애기
# sent.split(',').strip() - 오류: string.strip() 이므로 에러
```

```python
# 해결
[word for word in sent.split(',')]
```

```python
sent_tokens = [word for word in sent.split(',')]
sent_tokens
```

```python
#['I', 'have', 'a', 'car'] => 'I have a car'
'-'.join(sent_tokens) # '()'.join: ()로 연결해줌
```
