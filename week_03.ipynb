### 논리연산자
- 비교연산자가 여러 번 사용될 때
- and, or, not
    - A and B: A도 참이고, B도 참일 때 True
        - ex. A & B
    - A or B: A 혹은 B, 둘 중 하나만 참이면 True
        - ex. A | B
    - not (A): A가 False -> True, True -> False
- 비교연산자의 괄호는 가독성을 위해 사용 (optional, 생략가능)
- 파이썬에서만 가능한 표현: 0 < x < 10 식의 표현

### True, False
- "만약 a가 참이라면, ..."
-  a = True

```python
if a == True:
    print('something') # (x)
    
if a:
    print('something') # (o)
```

### in
- membership operator
- 어떤 요소가 안에 있는지 확인하고 싶을 때
- 비슷한 조건이 여러 번 반복될 때 사용 가능

```python
letter = 'k'

# letter이 모음인지 확인하고 싶다: a, e, i, o, u
# if letter = 'a' or letter == 'e', ... 일일이 확인 => 'in'을 사용하여 간소화 가능

if letter in ['a', 'e', 'i', 'o', 'u']: # 리스트 요소에 letter와 일치하는 것이 있는지 확인
    print("Yes")
else:
    print("No")
    
# if letter not in ~ : 
```

```python
word = 'abcde'

# word에 모음이 있는지 확인하는 코드 작성하기: 학우들 코드 점검
```

```python
if 'a' or 'e' or 'i'or 'o' or 'u' in word:
    print("Yes")
else:
    print("No")
```

```python
if word in ['a', 'e', 'i', 'o', 'u']: # No로 출력되는 이유: abcde == a, b, ... 이냐를 묻는 것이기 때문에 코드 작성 취지와 맞지 않음. 오답
    print("Yes")
else: print("No")
```

```python
vowel = 'a', 'e', 'i', 'o', 'u'
# type(vowel) => tuple

if vowel in ['a', 'b', 'c', 'd', 'e']: # tuple이 list에 있는지를 묻는 것. 자료형이 달라 x
    print('Yes')
else:
    print('No')
```

### 바다코끼리 연산자
- walrus operator
- 할당과 테스트를 동시에 하는 기능
- :=

```python
tweet_limit = 200
tweet_string = 'blah' * 50 # 공백 없이 스트링 4*50개
tweet_string
```

```python
diff = tweet_limit - len(tweet_string) # string(글자)의 길이가 limit(글자수 제한)을 넘는지 판별
if diff > 0:
    print("가능")
else: print("불가능")
```

```python
"""
diff = tweet_limit - len(tweet_string)
if diff > 0:
"""

# 바다코끼리 연산자 사용

if diff := tweet_limit - len(tweet_string) > 0:
        print("가능")
else: print("불가능")
```

### Quiz1.
- 윤년을 구하는 코드
    - 연도가 4로 나누어떨어지면 윤년
    - 100으로 나누어떨어지면 윤년x
    - 400으로 나누어떨어지면 윤년
    
- year을 입력받는다.
- '{ }년은 윤년입니다' / '{ }년은 윤년이 아닙니다.'

```python
year = int(input("연도를 입력하세요: "))

if year % 4 == 0 and year % 100 != 0 or year % 400 == 0:
    print(f"{year}년은 윤년입니다.")
else:
    print(f"{year}년은 윤년이 아닙니다.")
```

### Slicing
- 지정 범위의 문자열을 추출
- 문법: 대괄호 + 오프셋 이용 ['offset']
        - [start:end:step]
- 오프셋은 0부터 시작
- start 포함, end 불포함 (()보다 크거나 같고, ()보다 작다)
- 0의 경우 생략 가능
    - 예: ~4까지 출력하라 => [:5]
    - 처음부터 끝까지 2스텝씩 => [::2]
    - 처음부터 끝까지 역순으로 찍기 => [::-1]
- 오프셋은 거꾸로도 가능
- 0 1 2 . .
- h e l l o
- -5 ...  -1

```python
sent, sent[1] # 2번째 문자
```

```python
sent[3:4] # 4번째 문자
```

```python
# 처음부터 4까지
sent[0:5]
sent[:5]
```

```python
sent2 = 'hello world'
len(sent2) # len: length (공백까지 포함한 문자열 길이)
```

```python
# 처음부터 끝까지 2스텝씩 뛰어 출력하기
sent[::2]
```

```python
hello = 'hello'
hello[::-1] # 거꾸로 출력하기
```

```python
moby.tokens # 단어에 접근
'''

토큰: 프로그래밍에서 정의하는 단어
"I'm a boy." i'm, i am, i 'm
'''
```

```python
# 중복되는 단어들을 걸러내기
set(moby.tokens)
```

- set
    - {a, b, c, ...}
    - 중복 불가능
    - 수학의 집합과 동일한 개념
    - 순서 없음 (슬라이싱 불가)
- list
    - [a, b, c, ...]
    - 중복 가능
    - 순서 있음

```python
cars = ['kia', 'benz', 'porsche', 'kia']
set(cars) # set으로 감싸 형변환 -> 중복인 'kia' 사라짐
```

```python
# 중복 없는 단어 수
len(set(moby.tokens))
```

```python
# 어휘의 풍부성 측정: lexical diversity => 중복 없는 단어 수 / 단어 수
len(set(moby.tokens)) / len(moby.tokens)
```

```python
# 오름차순, 내림차순 정렬하기
sorted(cars)
sorted(cars, reverse = True)
```

```python
# mobydick
sorted(set(moby.tokens), reverse = True) [:10] # (슬라이싱으로 줄여주기)
```

- 출력:
    - "당신은 {}년 {}월 {}일 출생의 {남성/여성}입니다."
    - 휴대폰번호: {-} #하이픈으로 연결하기
    - 이메일 주소: {아이디}@xx.com 으로 출력

```python
regi = input('주민등록번호: ')
num = input('휴대폰 번호: ')
email = input('이메일 아이디: ')

if regi[6] == '1' or regi[6] == '3': #머지...왜 1 or 3이라하면 안되는거임
    sex = '남자'
elif regi[6] == '2' or regi[6] == '4':
    sex = '여자'
else:
    print('주민등록번호 뒷자리 1번째 숫자를 잘못 입력하셨습니다.')

print(f'당신은 {regi[0:2]}년 {regi[2:4]}월 {regi[4:6]}일 출생의 {sex}입니다.')
print(f'휴대폰 번호: {num[0:3]}-{num[3:7]}-{num[7:11]}')
print(f'이메일 주소: {email}@naver.com')
```
